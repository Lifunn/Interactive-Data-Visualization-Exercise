```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import RangeTool
from bokeh.sampledata.stocks import MSFT
import pandas as pd

output_notebook()
```

```python
df = pd.DataFrame(MSFT)
df["date"] = pd.to_datetime(df["date"])
```

```python
df.head()
```

```python
x = df["date"]
y = df["close"]
```

```python
p = figure(
    height=300,
    width=800,
    tools="xpan",
    toolbar_location=None,
    x_axis_type="datetime",
    x_axis_location="above",
    background_fill_color="#efefef",
    x_range=(x.iloc[0], x.iloc[200])
)
p.line(x, y, color="navy")
p.yaxis.axis_label = "High Price"
p.title.text = "Time Series Stock Data"
```

```python
select = figure(
    height=130,
    width=800,
    y_range=p.y_range,
    x_axis_type="datetime",
    y_axis_type=None,
    tools="",
    toolbar_location=None,
    background_fill_color="#efefef"
)
select.line(x, y, color="navy")
select.ygrid.grid_line_color = None

range_tool = RangeTool(x_range=p.x_range)
range_tool.overlay.fill_color = "green"
range_tool.overlay.fill_alpha = 0.3

select.add_tools(range_tool)
```

```python
show(p)
show(select)
```
