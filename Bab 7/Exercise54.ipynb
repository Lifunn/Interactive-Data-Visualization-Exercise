```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
w = pd.read_csv('/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/weight.csv')
w.head()
```

```python
w.describe()
```

```python
doc_w = w[w['weight']!= 0]
doc_w.describe()
```

```python
fig1 = px.box(doc_w, x = 'sex', y = 'weight', notched = True)
fig1.show()
```

```python
mean_w = doc_w['weight'].mean()
```

```python
w_new = w.replace({'weight': {0:mean_w}})
```

```python
w_new.describe()
```

```python
fig2 = px.box(w_new, x = 'sex', y = 'weight', notched = True)
fig2.show()
```
