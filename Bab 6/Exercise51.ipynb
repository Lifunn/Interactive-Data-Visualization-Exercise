```python
import pandas as pd

us_airports_df = pd.read_csv(
    '/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/airports.csv'
)

us_airports_df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Tambahkan trace scattergeo
fig.add_trace(
    go.Scattergeo(
        locationmode='USA-states',
        lon=us_airports_df['LONGITUDE'],
        lat=us_airports_df['LATITUDE'],
        hoverinfo='text',
        text=us_airports_df['AIRPORT'],
        mode='markers',
        marker=dict(size=5, color='black')
    )
)

# Update layout peta
fig.update_layout(
    title_text='Airports in the USA',
    showlegend=False,
    geo=go.layout.Geo(
        scope='usa'
    )
)

fig.show()
```

```python
new_year_2015_flights_df = pd.read_csv('/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/new_year_day_2015_delayed_flights.csv')
new_year_2015_flights_df.head()
```

```python
# Merge DataFrames on origin airport codes
new_year_2015_flights_df = new_year_2015_flights_df.merge(
    us_airports_df[['IATA_CODE','LATITUDE','LONGITUDE']],
    left_on='ORIGIN_AIRPORT',
    right_on='IATA_CODE',
    how='inner'
)

# Drop the duplicate column containing airport code
new_year_2015_flights_df.drop(columns=['IATA_CODE'], inplace=True)

# Rename latitude and longitude columns to reflect origin airport
new_year_2015_flights_df.rename(
    columns={
        "LATITUDE": "ORIGIN_AIRPORT_LATITUDE",
        "LONGITUDE": "ORIGIN_AIRPORT_LONGITUDE"
    },
    inplace=True
)

# Preview the first 5 rows
new_year_2015_flights_df.head()
```

```python
# Merge the DataFrames on destination airport codes
new_year_2015_flights_df = new_year_2015_flights_df.merge(
    us_airports_df[['IATA_CODE','LATITUDE','LONGITUDE']],
    left_on='DESTINATION_AIRPORT',
    right_on='IATA_CODE',
    how='inner'
)

# Drop the duplicate column containing airport code
new_year_2015_flights_df.drop(columns=['IATA_CODE'], inplace=True)

# Rename latitude and longitude columns to reflect destination airport
new_year_2015_flights_df.rename(
    columns={
        'LATITUDE': 'DESTINATION_AIRPORT_LATITUDE',
        'LONGITUDE': 'DESTINATION_AIRPORT_LONGITUDE'
    },
    inplace=True
)

# Preview the first 5 rows
new_year_2015_flights_df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Tambahkan garis penerbangan dari origin ke destination
for i in range(len(new_year_2015_flights_df)):
    fig.add_trace(
        go.Scattergeo(
            locationmode='USA-states',
            lon=[
                new_year_2015_flights_df['ORIGIN_AIRPORT_LONGITUDE'][i],
                new_year_2015_flights_df['DESTINATION_AIRPORT_LONGITUDE'][i]
            ],
            lat=[
                new_year_2015_flights_df['ORIGIN_AIRPORT_LATITUDE'][i],
                new_year_2015_flights_df['DESTINATION_AIRPORT_LATITUDE'][i]
            ],
            mode='lines',
            line=dict(width=1, color='red')
        )
    )

# Update layout peta
fig.update_layout(
    title_text='Delayed flight on Jan 1, 2015 in USA',
    showlegend=False,
    geo=go.layout.Geo(scope='usa')
)

fig.show()
```
