```python
import pandas as pd
import plotly.express as px
```

```python
production_df = pd.read_csv("/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/renewable-energy-consumption-by-country.csv")
production_df.head()
```

```python
production_df = production_df.sort_values(by='Year')
production_df.head()
```

```python
col_name = [c for c in production_df.columns if 'electricity' in c.lower() or 'total' in c.lower()][0]
col_name
```

```python
fig = px.choropleth(
    production_df,
    locations="Code",
    color=col_name,
    hover_name="Country",
    animation_frame="Year",
    color_continuous_scale="Greens",
    title="Renewable energy production across the world (% of electricity production)"
)
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='natural earth')
)
fig.show()
```

```python
consumption_df = pd.read_csv("/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/renewable-energy-consumption-by-country.csv")
consumption_df.head()
```

```python
consumption_long = pd.melt(
    consumption_df,
    id_vars=['Country', 'Code', 'Year'],
    var_name='Energy Source',
    value_name='Consumption (terawatt-hours)'
)
```

```python
consumption_long = consumption_long.sort_values(by="Year")
```

```python
fig2 = px.choropleth(
    consumption_long,
    locations="Code",
    color="Consumption (terawatt-hours)",
    hover_name="Country",
    animation_frame="Year",
    color_continuous_scale="YlGnBu",
    title="Renewable energy consumption across the world (Terawatt-hours)"
)

fig2.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='natural earth')
)
fig2.show()
```
