```python
import pandas as pd

walmart_loc_df = pd.read_csv(
    '/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/1962_2006_walmart_store_openings.csv'
)

walmart_loc_df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Scattergeo(
            lon=walmart_loc_df['LON'],  # kolom longitude
            lat=walmart_loc_df['LAT'],  # kolom latitude
            text=walmart_loc_df['STREETADDR'],  # alamat yang ditampilkan saat hover
            mode='markers',  # titik marker untuk setiap lokasi
            marker=dict(
                size=4,
                color='blue',
                opacity=0.7
            )
        )
    ]
)

fig.update_layout(
    title='Walmart stores across world',
    geo_scope='usa',  # fokus peta ke wilayah AS
)

fig.show()
```
