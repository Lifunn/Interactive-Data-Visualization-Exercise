```python
import pandas as pd
```

```python
internet_usage_df = pd.read_csv('/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/share-of-individuals-using-the-internet.csv')
internet_usage_df.head()
```

```python
import plotly.express as px

fig = px.choropleth(
    internet_usage_df,
    locations="Code",
    color="Individuals using the Internet (% of population)",
    hover_name="Country",
    animation_frame="Year",
    color_continuous_scale=px.colors.sequential.Plasma
)

fig.update_layout(
    title_text='Internet usage across the world (% population)',
    geo=dict(
        projection={'type': 'natural earth'}
    )
)

fig.show()
```

```python
internet_usage_df.sort_values(by=["Year"],inplace=True)
internet_usage_df.head()
```

```python
import plotly.express as px

fig = px.choropleth(
    internet_usage_df,
    locations="Code",  # kolom ISO Alpha-3 negara
    color="Individuals using the Internet (% of population)",  # kolom nilai warna
    hover_name="Country",  # nama negara saat hover
    animation_frame="Year",  # animasi berdasarkan tahun
    color_continuous_scale=px.colors.sequential.Plasma  # skema warna
)

# Mengatur layout peta
fig.update_layout(
    title_text='Internet usage across the world (% population)',
    geo=dict(projection_type='natural earth')  # tipe proyeksi peta
)

fig.show()
```
