```python
import pandas as pd

walmart_loc_df = pd.read_csv(
    '/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/1962_2006_walmart_store_openings.csv'
)

walmart_loc_df.head()
```

```python
walmart_stores_by_state = walmart_loc_df.groupby('STRSTATE').count()['storenum'].reset_index().rename(columns={'storenum':'NUM_STORES'})
walmart_stores_by_state.head()
```

```python
import plotly.express as px

fig = px.scatter_geo(
    walmart_stores_by_state,
    locations="STRSTATE",         # kolom yang berisi kode state (misal: 'CA', 'TX', 'NY')
    size="NUM_STORES",            # kolom yang berisi jumlah toko Walmart per state
    locationmode='USA-states',    # mode lokasi khusus untuk US
    hover_name="STRSTATE",        # tampilkan kode state saat hover
    size_max=45                   # ukuran maksimum bubble
)

# Mengatur layout peta
fig.update_layout(
    title_text='Walmart stores across states in the US',  # judul peta
    geo_scope='usa'  # batasi tampilan hanya wilayah AS
)

fig.show()
```

```python
internet_users_df = pd.read_csv('/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/number-of-internet-users-by-country.csv')
internet_users_df.head()
```

```python
internet_users_df.sort_values(by=['Year'],inplace=True)
internet_users_df.head()
```

```python
import plotly.express as px

fig = px.scatter_geo(
    internet_users_df.query("Year == 2016"),
    locations="Code",  # name of column indicating country-codes
    size="Number of internet users (users)",  # name of column by which to size the bubble
    hover_name="Country",  # name of column to be displayed while hovering over the map
    size_max=80,  # parameter to scale all bubble sizes
    color="Number of internet users (users)",  # tambahkan color agar warna muncul
    color_continuous_scale=px.colors.sequential.Plasma  # color scale
)

fig.update_layout(
    title_text='Internet users across the world - 2016',  # add a title text for the plot
    geo=dict(
        projection_type='natural earth'  # set projection style for the plot
    )
)

fig.show()
```
