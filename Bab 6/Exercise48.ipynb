```python
import pandas as pd

df = pd.read_csv(
    '/content/drive/MyDrive/Visualisasi Informasi/Interactive-Data-Visualization-with-Python/datasets/us_state_population.tsv',
    sep='\t'
)

df.head()
```

```python
df = pd.melt(df, id_vars=['State', 'Code'], var_name="Year", value_name="Population")
df.head()
```

```python
import plotly.graph_objects as go
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Choropleth(
            locations=df['Code'],              # Kode untuk tiap state (misal: 'CA', 'TX', 'NY', dst)
            z=df['Population'].astype(int),    # Nilai yang divisualisasikan (populasi)
            locationmode='USA-states',         # Mode lokasi (khusus peta USA)
            colorscale='Blues',                # Skema warna
            colorbar_title="Population"        # Judul colorbar
        )
    ]
)

# Tambahkan layout peta (agar peta fokus ke USA)
fig.update_layout(
    title_text='US States Population',
    geo_scope='usa'  # membatasi tampilan ke wilayah USA
)

fig.show()
```
